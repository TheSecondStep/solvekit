## Developer Guide

If you are new to using `nbdev` here are some useful pointers to get you started.

### Install solvekit in Development mode

```sh
# make sure solvekit package is installed in development mode
$ pip install -e .

# make changes under nbs/ directory
# ...

# compile to have changes apply to solvekit
$ nbdev_prepare
```

## Usage

### Installation

Install latest from the GitHub [repository][repo]:

```sh
$ pip install git+https://github.com/TheSecondStep/solvekit.git
```

or from [conda][conda]

```sh
$ conda install -c TheSecondStep solvekit
```

or from [pypi][pypi]


```sh
$ pip install solvekit
```
